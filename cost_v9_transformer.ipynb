```python
import pandas as pd
import numpy as np
import warnings


warnings.filterwarnings('ignore', 'PerformanceWarning')
```

```python
data = pd.read_excel("cost_v9.xlsx", sheet_name=1)
data = data[['Name EN', 'Type', 'Lifeform', 'metal base cost',
       'crystal base cost', 'deut base cost', 'energy base cost',
       'metal increase factor', 'crystal increase factor',
       'deut increase factor', 'energy increase factor', ]]

# on vire les na

data.fillna(0, inplace=True)

# Traduction

from google_trans_new import google_translator  
from tqdm import tqdm

translator = google_translator()

translations = {}

for word in tqdm(data['Name EN']):
    translations[word] = translator.translate(word, lang_tgt="fr")
 
# quelques modifs pour que ça corresponde aux jeux :

translations['Biosphere Farm'] = "Ferme biosphérique" 
translations['Neuro-Calibration Centre'] =  "Centre de neurocalibrage"
translations['Food Silo'] = "Réserve alimentaire"    
translations['Metropolis'] = "Métropolis"
translations['Fusion-Powered Production'] = "Extraction par fusion"
translations['Skyscraper\n'] = "Tour d'habitation"
translations['Intergalactic Envoys'] = "Ambassadeurs intergalactiques"
translations['High-Performance Extractors'] = "Extracteurs à haute performance"
translations['Fusion Drives'] = 'Moteurs à fusion'
translations['Sanctuary\n'] = "Refugium"

translations
```

```python
data['Name FR'] = data['Name EN'].replace(translations)
```

```python
# Ajout jusqu'au level 30
```

```python
for i in range(1,60):
    data[f'metal cost {i}'] = np.floor(data['metal base cost'] * pow(data['metal increase factor'], i-1)*i)/1000
    data[f'crystal cost {i}'] = np.floor(data['crystal base cost'] * pow(data['crystal increase factor'], i-1)*i)/1000
    data[f'deut cost {i}'] = np.floor(data['deut base cost'] * pow(data['deut increase factor'], i-1)*i)/1000
    data[f'energy cost {i}'] = np.floor(data['energy base cost'] * pow(data['energy increase factor'], i-1)*i)/1000
```

```python
data.to_excel("data_cost.xlsx", index=False)
```
